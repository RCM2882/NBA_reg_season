# Nba Regular Season 2018-19 Data Challenge

Your task will be to take the dataset given, and create an analysis answering the following 10 questions. This project will again test your knowledge of pandas in order to find the answers needed given the data you are presented with.

# What was the average age of player in the league?

```python
import pandas as pd
season_stats = pd.read_csv('nbastats2018-2019.csv', sep = ',')

avg_age = season_stats['Age'].mean()

print("Average age for players in league:")
print(avg_age)
```

# What player scored the most points?

```python

most_points = season_stats.sort_values(by='Points', ascending = False, kind = 'quicksort')

print('Top scorer for 2018-2019 season was James Harden')
most_points.head()
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python

most_blocks = season_stats.sort_values(by='Blocks', ascending = False, kind = 'quicksort')

print('The player that averaged the most blocks per game, during the 2018-2019 season was Myles Turner')
print('He plays Center for the Indiana Pacers ')
most_blocks.head()
```

# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
season_stats['Glory'] = season_stats['PER'] + season_stats['WS']

contenders = season_stats.sort_values(by='Glory', ascending = False )

print("I would've predicted that the Golden State Warriors had the best chance to win a title,")
print("because they had the most players in the top twenty of Win Shares + Player Efficiency Rating")
print("(Glory) with a minimum 30 games played during the regular season")

contenders.head(20)
```

# What player had the best 3-pt percentage? 

```python
best_3pt_prct = season_stats.sort_values(by='FG3%', ascending = False)

print('The player with the best 3_pt% for 2018-2019 season, with a minimum 30 games played')
print('is Domantas Sabonis')
best_3pt_prct.head(10)
```

# Who played the most minutes during the season

```python
most_min = season_stats.sort_values(by='MP', ascending = False)

print('The player that averaged the most minutes played for the 2018-2019 season was')
print('a tie between Paul George and Bradley Beal')
most_min.head(10)
```

# What player given their player effiecency rating was the clutchest during the season?

```python
player_eff = season_stats.sort_values(by='PER', ascending = False)

print('The player with the best Player Efficiency Rating for 2018-2019 season,')
print('with a minimum 30 games played was Giannis Antetokounmpo')
player_eff.head(10)
```

# What team had the youngest roster?

```python
avg_age_by_roster = season_stats.groupby('Team')['Age'].mean()

youngest_roster = avg_age_by_roster.idxmin()

print(f'The team with the youngest roster for the 2018-2019 season was: {youngest_roster}')
```

# Who is the highest paid player during the seasion?

```python
highest_paid = season_stats.sort_values(by='Salary', ascending = False)

print('The player with a minimum 30 games played and the highest Salary for 2018-2019 season was Dwight Powell')

highest_paid.head()
```

# At the end of a game, who WOULDN'T you want on the Free Throw Line?

```python
worst_ft = season_stats.sort_values(by='FT%', ascending = True)

print('The player, with a minimum 30 games played, you wouldn"t want at the')
print('Free Throw Line at the end of the game is Jacob Evans')
worst_ft.head()
```
